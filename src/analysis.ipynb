```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open("./analysis/powerLaw.json") as json_file:
    pl = json.load(json_file)
    
x = range(len(pl.keys()))
plt.figure(dpi=100)

for node_id in pl["0"].keys():
    y = []
    for time in pl.keys():
        y.append(pl[time][node_id])
    #plt.plot(x, y, label=str(node_id))
    plt.plot(x, y)

plt.xlabel('Time')
plt.ylabel('Influence') 
plt.title('Influence of Initial Nodes over time t')

#plt.legend()  
plt.savefig("pl")
# plt.show()
```

```python
with open("./analysis/density.json") as json_file:
    density = json.load(json_file)
    
x = np.arange(0, len(density["density"]), 1)

y1 = density["density"]
y2 = density["nodes"]
y3 = density["edges"]
y4 = density["degree"]

plt.figure(dpi=100)
plt.plot(x, y2, label="nodes")
plt.plot(x, y3, label="edges")
plt.legend()
plt.xlabel('time')
plt.ylabel('Count') 
plt.title('No. of nodes and edges over time t')
plt.savefig("ne1")

plt.figure(dpi=100)
plt.plot(x, y1, label="Density")
plt.legend()    
plt.xlabel('time')
plt.ylabel('Density') 
plt.title('Density over time t')
plt.savefig("density")

plt.figure(dpi=100)
plt.plot(x, y4, label="avg_degree")
plt.legend()
plt.xlabel('time')
plt.ylabel('Average Degree') 
plt.title('Average Node Degree over time t')
plt.savefig("degree")
# plt.show()
```

```python
with open("./analysis/degreeCentrality.json") as json_file:
    degCentralilty = json.load(json_file)
    
    
plt.figure(dpi=100)
plt.hist(degCentralilty.values())
plt.xlabel('Degree Centrality')
plt.ylabel('Nodes') 
plt.title('Degree Centrality Histogram')
#plt.show()
plt.savefig("dc1")
        
X = []
Y = []
for ind in degCentralilty.keys():
    X.append(int(ind))
    Y.append(degCentralilty[ind])
    
plt.figure(dpi=100)
plt.plot(X, Y)
plt.axhline(np.mean(Y),color='red', linestyle='--')
plt.xlabel('Nodes')
plt.ylabel('Degree Centrality') 
plt.title('Degree Centrality')
plt.savefig("dc2")
#plt.show()


with open("./analysis/twitterdegreeCentrality.json") as json_file:
    twitterdegCentralilty = json.load(json_file)
    
    
plt.figure(dpi=100)
plt.hist(twitterdegCentralilty.values())
plt.xlabel('Degree Centrality')
plt.ylabel('Nodes') 
plt.title('Twitter Degree Centrality Histogram')
#plt.show()
plt.savefig("dc3")

node_id = []
for nodeID in twitterdegCentralilty.keys():
    node_id.append(int(nodeID))
    
node_id = sorted(node_id)
        
X = []
Y = []
for ind in node_id:
    X.append(int(ind))
    Y.append(twitterdegCentralilty[str(ind)])
    
plt.figure(dpi=100)
plt.plot(X, Y)
plt.axhline(np.mean(Y),color='red', linestyle='--')
plt.xlabel('Nodes')
plt.ylabel('Degree Centrality') 
plt.title('Twitter Degree Centrality')
plt.savefig("dc4")
```
